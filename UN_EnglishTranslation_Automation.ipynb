```python
# This code shows how to translate to English as an automatic process with Python

# 0. install the library

# 1. import file

# 2. run the self-defined function cell (translate_in_english)

# 3. by using the apply(translate_in_english), the text will be translated and a new column, translated.
```

```python
# import the file to use
# if the file is saved as csv, use pd.read_csv('file path and name')
# this code assumes that the file you want to load is written in excel (xlsx format)
# shows the first 5 lines of code to see whether the file is correctly imported
# if there's a blank cell, the function will make an error. So, use fillna() function.

import pandas as pd

df = pd.read_excel('/content/BOS_RISKS (1).xlsx')
df["RISK_DESCRIPTION"].fillna("", inplace=True)
df.head()
```

```python
# self defined function to use google translator
# this code translated any written text to English by google translator

from deep_translator import GoogleTranslator

def translate_in_english(text):
  translated = GoogleTranslator(source="auto", target="en").translate(text)
  return translated
```

```python
# Be patinet. The function takes some time
# this code will be updated to reduce the processing time later

df["translated"] = df["RISK_DESCRIPTION"].apply(translate_in_english)

df.head()
```
